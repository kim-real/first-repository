```python
# 오차 측정 RMSE
def RMSE(a, b):
    mse = ((a - b) ** 2).mean()  # 두 값의 차이의 제곱의 평균
    rmse = mse ** 0.5        # MSE의 제곱근
    return rmse
print("슝=3")
```

```python
#예측값과 정답값에 대한 차이를 계산하는 함수를 손실함수(loss function), 또는 비용함수(cost function) 라고 부릅니다.
#손실값(loss)이 크다는 것은 모델이 현재 정답과 굉장히 먼 예측을 하고 있다는 뜻이고, 
#손실값이 작다는 것은 모델이 정답에 가까운, 올바른 예측을 하고 있다는 것으로 해석할
def loss(x, w, b, y):
    predictions = model(x, w, b)
    L = RMSE(predictions, y)
    return L
print("슝=3")
```

#Gradient Descent Method, 즉 경사하강법
#오른쪽으로 갈수록 손실함수값이 점점 작아지기 때문에 그 점에서 그린 접선은 오른쪽 아래로 향하는 방향 이고, 그 기울기는 음수 입니다.
#오른쪽 점은 자기 자신의 위치보다 왼쪽으로 갈수록 손실함수값이 점점 작아지기 때문에 
#그 점에서 그린 접선은 왼쪽 아래로 향하는 방향이고, 그 기울기는 양수 
#현재 w에서의 기울기를 구한다.
#기울기가 음수라면, 현재 w를 키운다.
#기울기가 양수라면, 현재 w를 줄인다
#w new w =w−ηg
#냥 w에서 기울기를 빼 주면 바로 우리가 원하는대로 동작할 
#g가 음수라면 w-gw−g 식에서 우리는 ww 에서 음수를 빼주게 되므로 ww는 커집니다.
#반면, gg가 양수라면 w-gw−g 식에서 우리는 ww 에서 양수를 빼주게 되므로 ww는 작아집니다.
#η 는 고정되어 있는 상수로써 ww를 변화시키는 양을 조절하는 역

```python
def gradient(x, w, b, y):
    dw = (loss(x, w + 0.0001, b, y) - loss(x, w, b, y)) / 0.0001
    db = (loss(x, w, b + 0.0001, y) - loss(x, w, b, y)) / 0.0001
    return dw, db
print("슝=3")
```

```python
'''#모델이 스스로 학습해나가는 파라미터 가 아니라, 사람이 직접 사전에 정하고 시작해야 하는 파라미터 를 하이퍼 파라미터 
입력 데이터 x와 정답 데이터 y 준비
초기 w, b를 랜덤하게 선택
현재의 w와 b로 모델이 prediction을 예측하고, 그 값과 y 값을 비교해서 손실함수값 계산
손실함수값을 계산함과 동시에 그 점에서의 기울기, gradient 계산
계산된 gradient를 활용해서 w'=w-\eta gw 
′
 =w−ηg , b'=b-\eta gb 
′
 =b−ηg 로 업데이트
3~5 번의 과정을 손실함수가 충분히 줄어들 때까지 반복
'''
```

```python
LEARNING_RATE = 1
print("슝=3")
```

```python
x = macbook["used_years"].values
y = macbook["price"].values
print("슝=3")
```

```python
w = 3.1
b = 2.3
w, b
```

```python
#손실함수값이 단계별로 얼마 정도인지를 저장할 losses라는 빈 리스트
losses = []
print("슝=3")
```

```python
for i in range(1, 2001):
    dw, db = gradient(x, w, b, y)   # 3, 4번: 모델이 prediction을 예측하고, 손실함수값을 계산함과 동시에 기울기 계산
    w -= LEARNING_RATE * dw         # 5번: w = w - η * dw 로 업데이트
    b -= LEARNING_RATE * db         # 5번: b = b - η * db 로 업데이트 
    L = loss(x, w, b, y)            # 현재의 loss 값 계산
    losses.append(L)                # loss 값 기록
    if i % 100 == 0:
        print('Iteration %d : Loss %0.4f' % (i, L))
```

```python
#loss 크래프
plt.plot(losses)
plt.show()
```

```python
#rendom prepare
import numpy as np
W = np.random.rand(12)
b = np.random.rand()
print("슝=3")
```

```python
def model(X, W, b):
    predictions = 0
    for i in range(12):
        predictions += X[:, i] * W[i]
    predictions += b
    return predictions
print("슝=3")
```

```python
def MSE(a, b):
    mse = ((a - b) ** 2).mean()  # 두 값의 차이의 제곱의 평균
    return mse
print("슝=3")
```

```python
def loss(X, W, b, y):
    predictions = model(X, W, b)
    L = MSE(predictions, y)
    return L
print("슝=3")
```

```python
'''우리가 손실함수를 미분하는 이유는 최적의 매개변수, ww와 bb에 가까워지는 방향을 알기 위함이었다.
특정 순간에 기울기가 양수라면 ww를 더 작게 줄이고, 기울기가 음수라면 ww를 더 크게 늘린다.
즉, 우리는 w' = w - \eta gw 
′
 =w−ηg 공식을 활용해서 변수를 업데이트하는데, 이 때 기울기는 변수 업데이트의 방향을 결정하는 역할을 한다.
따라서 손실함수는 언제나 내가 최적을 찾고자 하는 매개변수, 혹은 가중치에 대해 미분해야 한다.
현재 모델에는 w_1w 
1
​
  부터 w_{12}w 
12
​
 까지 12개의 ww 와, 1개의 bb가 있으므로 우리는 총 13개의 변수에 대해 미분을 해야 한다.
  (y_{i, true} - y_{i, pred})^2(y 
i,true
​
 −y 
i,pred
​
 ) 
2
  을 w_iw 
i
​
 로 미분'''
```

```python
def gradient(X, W, b, y):
    # N은 가중치의 개수
    N = len(W)
    
    # y_pred 준비
    y_pred = model(X, W, b)
    
    # 공식에 맞게 gradient 계산
    dW = 1/N * 2 * X.T.dot(y_pred - y)
        
    # b의 gradient 계산
    db = 2 * (y_pred - y).mean()
    return dW, db
print("슝=3")
```

```python
dW, db = gradient(X, W, b, y)
print("dW:", dW)
print("db:", db)
```

```python
LEARNING_RATE = 0.0001
```

```python
losses = []

for i in range(1, 1001):
    dW, db = gradient(X_train, W, b, y_train)
    W -= LEARNING_RATE * dW
    b -= LEARNING_RATE * db
    L = loss(X_train, W, b, y_train)
    losses.append(L)
    if i % 10 == 0:
        print('Iteration %d : Loss %0.4f' % (i, L))
```

```python
import matplotlib.pyplot as plt
plt.plot(losses)
plt.show()
```

```python
#(5) 모델 준비하기
##w,b준비
import numpy as np
w=0
b=0
W = np.random.rand(10)
b = np.random.rand()
```

```python
#(5) 모델 준비하기
#모델정의
def model(X, W, b):
    predictions = 0    #?????? y_pred
    for i in range(10):
        predictions += X[:, i] * W[i]
    predictions += b
    return predictions
```

```python
#(6) 손실함수 loss 정의하기
def loss(X, W, b, y):
    predictions = model(X, W, b)
    L = MSE(predictions, y)
    return L
```

```python
def MSE(a, b):
    mse = ((a - b) ** 2).mean()  # 두 값의 차이의 제곱의 평균
    return mse
```

```python
#(7) 기울기를 구하는 gradient 함수
#gradient 함수
def gradient(X, W, b, y):

# N은 가중치의 개수
    N = len(W)
    
    # y_pred 준비
    y_pred = model(X, W, b)
    
    # 공식에 맞게 gradient 계산
    dW = 1/N * 2 * X.T.dot(y_pred - y)
        
    # b의 gradient 계산
    db = 2 * (y_pred - y).mean()
    return dW, db
```

```python
#(8) 하이퍼 파라미터인 학습률 설정하기
#LEARNING_RATE   ( mse가 높으면 여러가지로 변경해서)
LEARNING_RATE = 0.0000000000001
```

```python
#(9) 모델 학습하기
#모델 학습하기
from sklearn.metrics import mean_squared_error
losses = []

for i in range(1, 1001):
    dW, db = gradient(X_train, W, b, y_train)
    W -= LEARNING_RATE * dW
    b -= LEARNING_RATE * db
    L = loss(X_train, W, b, y_train)
    losses.append(L)
    if i % 10 == 0:
        print('Iteration %d : Loss %0.4f' % (i, L))
```

```python
#loss 크래프
plt.plot(losses)
plt.show()
```

```python
#(10) test 데이터에 대한 성능 확인하기
prediction = model(X_test, W, b)
mse = loss(X_test, W, b, y_test)
mse
```
