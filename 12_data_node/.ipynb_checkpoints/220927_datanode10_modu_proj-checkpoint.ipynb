```python
#3-1) 데이터 현황 분석 .. 준비 ... 데이터(sales_data)를 temp_var로 정의하기
temp_var = sales_data
```

```python
#3-1) 데이터 현황 분석 .. 준비 ... date열을 만들고, sales_date의 결제일시 중 앞에서 7번째 값까지 할당
temp_var['Month'] = temp_var['결제일시'].str[:7]
```

```python
#3-1) 데이터 현황 분석 .. 준비 ...결제금액 int 컬럼 변환
temp_var['결제금액'] = temp_var.결제금액.str.replace(',', '').astype('int64')
```

```python
#3-2) 월별 매출액 추세 ...  결제금액 date 기준으로 집계하기 .. 2021-2,2022-04 data 삭제
temp_var['결제금액'].groupby(temp_var['Month']).sum()
```

```python
# 집계 결과를 시각화하기
plt.figure(figsize=(16,8))
temp_data = temp_var['결제금액'].groupby(temp_var['Month']).sum()
ax = plt.bar(temp_data.index, temp_data)
plt.ticklabel_format(axis='y',  style = 'plain')
plt.bar_label(ax, labels=[f'{x:,.0f}' for x in ax.datavalues], label_type='center')
plt.title('sales by Month')
plt.show()
```

```python
# temp_var의 상품명을 value_counts를 이용해 집계하기
# 주의! value_counts로 바로 집계하면 판매취소와 결제취소가 함께 집계됩니다.
# 따라서 '판매수량' 칼럼 값이 '-1'인 데이터를 제외하고 집계해야 합니다.
# 힌트 코드 : temp_var[temp_var['판매수량']==-1]
temp_var['결제금액'].groupby(temp_var['Month']).sum()
```

```python
#2-2) item data  전처리 판매금액 int 컬럼 추가
item_data['price'] = item_data.판매금액.str.replace('원', '')
item_data['price'] = item_data.price.str.replace(',', '').astype('int64')
```

```python
#2-3) item_entry_data  전처리 결제금액 int 컬럼 추가
item_entry_data['fee'] = item_entry_data.판매금액.str.replace('원', '')
item_entry_data['fee'] = item_entry_data.fee.str.replace(',', '').astype('int64')
```
